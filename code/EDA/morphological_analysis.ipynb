```python
from konlpy.tag import Mecab, Okt, Kkma, Komoran, Hannanum
from kiwipiepy import Kiwi
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
mecab = Mecab()
okt = Okt()
kkma = Kkma()
komoran = Komoran()
hannanum = Hannanum()
kiwi = Kiwi()
```

```python
train_path = "../../data/train.csv"
df_train = pd.read_csv(train_path)
```

```python
dialogue = df_train["dialogue"][0]
```

```python
mecab_pos = mecab.pos(dialogue)
okt_pos = okt.pos(dialogue)
kkma_pos = kkma.pos(dialogue)
komoran_pos = komoran.pos(dialogue)
hannanum_pos = hannanum.pos(dialogue)
kiwi_pos = kiwi.tokenize(dialogue)

print("== Mecab ==")
print(mecab_pos)
print("== Okt ==")
print(okt_pos)
print()
print("== Kkma ==")
print(kkma_pos)
print()
print("== Komoran ==")
print(komoran_pos)
print()
print("== Hannanum ==")
print(hannanum_pos)
print()
print("== Kiwi ==")
print(kiwi_pos)
```
